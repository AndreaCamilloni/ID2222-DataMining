```python
# Read dat file of generated transactions (baskets) of hashed items into list
with open('T10I4D100K.dat', 'r') as f:
    baskets = f.readlines()

# Remove whitespace characters like `\n` at the end of each line
baskets = [x.strip() for x in baskets]

# Convert list of baskets into pandas dataframe
df = pd.DataFrame(baskets)

#df = pd.read_csv('T10I4D100K.dat', sep='\n', header=None)
```

```python
# Itemset
itemset = set()

for basket in df[0]:
    itemset = itemset.union(basket)
```

```python
item_counts = {}

for item in itemset:
    item_counts[item] = 0
    for basket in df[0]:
        if item in basket:
            item_counts[item] += 1
```

```python
# frequent itemset
frequent_itemset = set()
s= len(df[0])*0.01 #1 percent of the total number of baskets

for item in itemset:
    if item_counts[item] >= s:
        frequent_itemset.add(item)

len(frequent_itemset)
```

```python
#frequent_pairs = []
pair_counts = {}

for basket in df[0]:
    frequent_pairs = basket.intersection(frequent_itemset)
   
    for item1 in frequent_pairs:
        for item2 in frequent_pairs:
            if item1 != item2:
                pair_counts[(item1, item2)] = pair_counts.get((item1, item2), 0) + 1

# Frequent pairs
frequent_pairs = set()

for pair in pair_counts:
    if pair_counts[pair] >= s:
        frequent_pairs.add(pair)

frequent_pairs
```

```python

triples_counts = {}

for basket in df[0]:
    for pair in frequent_pairs:
        if pair[0] in basket and pair[1] in basket:
            for item in basket:
                if item != pair[0] and item != pair[1]:
                    triples_counts[(pair[0], pair[1], item)] = triples_counts.get((pair[0], pair[1], item), 0) + 1
                    
# Frequent triples
frequent_triples = set()

for triple in triples_counts:
    if triples_counts[triple] >= s:
        frequent_triples.add(triple)

frequent_triples
```

```python
# Apriori algorithm given a set of baskets and a threshold
def apriori(df, threshold):
    # Itemset
    itemset = set()

    for basket in df[0]:
        itemset = itemset.union(basket)

    # Count the number of baskets each item appears in
    item_counts = {}

    for item in itemset:
        item_counts[item] = 0
        for basket in df[0]:
            if item in basket:
                item_counts[item] += 1

    # Frequent itemset
    frequent_itemset = set()

    for item in itemset:
        if item_counts[item] >= threshold:
            frequent_itemset.add(item)

    # Frequent pairs
    frequent_pairs = set()
    pair_counts = {}

    for basket in df[0]:
        frequent_pairs = basket.intersection(frequent_itemset)

        for item1 in frequent_pairs:
            for item2 in frequent_pairs:
                if item1 != item2:
                    pair_counts[(item1, item2)] = pair_counts.get((item1, item2), 0) + 1
    
    frequent_pairs = set()

    for pair in pair_counts:
        if pair_counts[pair] >= threshold:
            #print(pair)
            frequent_pairs.add(pair)
    
    
    # Frequent triples
    frequent_triples = set()
    triples_counts = {}

    for basket in df[0]:
        for pair in frequent_pairs:
            if pair[0] in basket and pair[1] in basket:
                for item in basket:
                    if item != pair[0] and item != pair[1]:
                        triples_counts[(pair[0], pair[1], item)] = triples_counts.get((pair[0], pair[1], item), 0) + 1

    for triple in triples_counts:
        if triples_counts[triple] >= threshold:
            frequent_triples.add(triple)
   

    return frequent_itemset,  frequent_pairs, frequent_triples
```
